2) Diplay the "ask" or "bid" price of an asset. Direction and asset name as parameters
def getDepth(direction='ask', pair='BTCUSD')

```python
def getDepth(direction, pair):
    if(direction == 'ask'):
        query2 = request_api_binance(path = "/api/v3/ticker/bookTicker?symbol=" + pair, headers = headers, params = params)
        ask_price_asset = query2["askPrice"]
        print(f"The ask price for asset {pair} is {ask_price_asset}")
    else:
        if(direction == 'bid'):
            query2 = request_api_binance(path = "/api/v3/ticker/bookTicker?symbol=" + pair, headers = headers, params = params)
            bid_price_asset = query2["bidPrice"]
            print(f"The bid price for asset {pair} is {bid_price_asset}")
        else:
            print('Paramètre de la fonction incorrecte')
    
```

```python
getDepth(direction='ask', pair='LTCBTC')
getDepth(direction='bid', pair='LTCBTC')
getDepth(direction='bids', pair='LTCBTC')
```
